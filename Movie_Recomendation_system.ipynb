#create movie matrics

```python
moviematrix=df3.pivot_table(index='user_id',columns='title',values='rating')
```

```python
moviematrix.head()
```

```python
moviematrix
```

```python
starwars_user_ratings = moviematrix['Star Wars (1977)']
```

```python
starwars_user_ratings
```

```python
similar_to_starwars = moviematrix.corrwith(starwars_user_ratings)
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars, columns =['Correlation'])
```

```python
corr_starwars
```

```python
corr_starwars.sort_values('Correlation', ascending = False).head(10)
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars, columns =['Correlation'])
corr_starwars.dropna(inplace = True)

corr_starwars.head()
```

```python
rating_df
```

```python
corr_starwars = corr_starwars.join(rating_df['number of rating'])
```

```python
corr_starwars
```

```python
corr_starwars.sort_values('Correlation', ascending = False)
```

```python
corr_starwars=corr_starwars[corr_starwars['number of rating']>1000]
```
