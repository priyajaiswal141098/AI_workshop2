```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python

# Apply LabelEncoder to each object column
for col in df.columns:
    if df[col].dtype == 'object':
      df[col] = le.fit_transform(df[col])
```

```python
df.info()
```

```python
df=df.drop('customerID',axis=1)
```

```python
df
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE(random_state=42)
```

```python
 X=df.drop('Churn',axis=1)
```

```python
y=df['Churn']
```

```python
X,y=sm.fit_resample(X,y)
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.25,
                                                    random_state=42)
```

```python
 X_test.shape
```

```python
 y_train.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(X_train,y_train)
```

```python
rfc.predict(X_test)
```

```python
accuracy=rfc.score(X_test,y_test)
```

```python
accuracy
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test,rfc.predict(X_test))
```

#Deep learning

```python
import keras
```

```python
model=keras.Sequential()
model.summary()
```

```python
X
```

```python
model.add(keras.Input(shape=(19,)) )
```

```python
model.add(keras.layers.Dense(32,activation='relu')) #relu for inpiut layer
```

```python
model.add(keras.layers.Dense(64,activation='relu')) #relu for inpiut layer
```

```python
model.add(keras.layers.Dense(1,activation='sigmoid')) #relu for inpiut layer
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
from gc import callbacks
callback=keras.callbacks.EarlyStopping(monitor='val_accuracy',
                                       patience=5,
                                       restore_best_weights=True)
```

```python
model.fit(X_train,
          y_train,
          epochs=100,
          batch_size=32,
          validation_data=(X_test,y_test),
          verbose="auto",
          callbacks=[callback])
```

```python
model.evaluate(X_test,y_test)
```
